## Exercise 1

Forces are applied at points A and B of the solid cast-iron bracket shown. Knowing that the bracket has a diameter of 0.8 in., determine the principal stresses and the maximum shearing stress at (a) point H, (b) point K.

![image1](8-41.png)

Please consider the following values for each:



```python
import numpy as np
import pandas as pd
lock = False
solve = True
if lock:
    classlist = pd.read_excel('classlist2021II.xlsx')

    A = np.random.choice(np.arange(1500, 3500, 6),20) # lb
    B = np.random.choice(np.linspace(1.5, 3.5, 6),20) # in
    C = np.random.choice(np.arange(500, 1000, 6),20) # lb
    D = np.random.choice(np.linspace(2.0, 4.0, 6),20) # in
    E = np.random.choice(np.linspace(0.5, 1.5, 6),20) # in
    var_d = {'A':A,'B':B,'C':C,'D':D, 'E':E}

    for key, item in var_d.items():
        classlist[key] = np.random.choice(item,16)

    classlist = classlist.round(3)
    classlist.to_csv('problem1.csv')
else:
    classlist = pd.read_csv('problem1.csv', index_col=[0])
```

```python
if solve:
    #Some common variables for all 
    c = 0.4 #in
    A = np.pi*c**2
    I = (np.pi*c**4)/4
    J = 2*I
    Q = (c**3)*2/3
    #Determining Mx
    classlist['Mx'] = (classlist['D']-classlist['E'])*classlist['C']
    #Determining Mz
    classlist['Mz'] = -classlist['B']*classlist['C']


    #At point H
    classlist['SigmaH'] = -classlist['A']/A + classlist['Mx']*c/I
    #Sigma average
    classlist['SigmaAveH'] = classlist['SigmaH']/2
    # Shear at Th
    classlist['TauH'] = classlist['Mz']*c/J
    # Radious
    classlist['Rh'] = np.sqrt(classlist['SigmaAveH']**2 + classlist['TauH']**2)
    # principal 1
    classlist['sH1'] = classlist['SigmaAveH'] + classlist['Rh']
    # principal 2
    classlist['sH2'] = classlist['SigmaAveH'] - classlist['Rh']
    # Tmax
    classlist['TmaxH'] =  (classlist['sH1'] + classlist['sH2']) /2

    # At point K
    classlist['SigmaK'] = -classlist['A']/A
    #Shear Tk
    classlist['TauK'] = -classlist['Mz']*c/J + classlist['C']*Q/(I*2*c)
    #Sigma average
    classlist['SigmaAveK'] = classlist['SigmaK']/2
    # Radious
    classlist['Rk'] = np.sqrt(classlist['SigmaAveK']**2 + classlist['TauK']**2)
    # principal 1
    classlist['sK1'] = classlist['SigmaAveK'] + classlist['Rk']
    # principal 2
    classlist['sK2'] = classlist['SigmaAveK'] - classlist['Rk']
    # Tmax
    classlist['TmaxK'] =  (classlist['sK1'] - classlist['sK2']) /2

    #classlist.iloc[student]
classlist
```

## Exercise 2

Three forces are applied to the bar shown. Determine the normal and shearing stresses at (a) point a, (b) point b, (c) point c.

![image2](8-47.png)

```python
if lock:
    classlist2 = pd.read_excel('classlist2021II.xlsx')

    F = np.random.choice(np.arange(50, 100, 6),20) # mm
    G = np.random.choice(np.arange(15,30, 6),20) # mm
    H = np.random.choice(np.arange(150, 300, 6),20) # mm
    I = np.random.choice(np.arange(600, 1000, 6),20) # I
    J = np.random.choice(np.arange(300,1000, 6),20) # N
    K = np.random.choice(np.arange(5, 30, 6),20) # KN
    var_d = {'F':F,'G':G,'H':H,'I':I, 'J':J, 'K':K}

    for key, item in var_d.items():
        classlist2[key] = np.random.choice(item,16)

    classlist2 = classlist2.round(3)
    classlist2.to_csv('problem2.csv')
else:
    classlist2 = pd.read_csv('problem2.csv', index_col=[0])
```

```python
if solve:
    classlist2['Area'] = classlist2['F']*32/(1000**2)
    classlist2['Iz'] = (classlist2['F']*32**3)/(12*1000**4)
    classlist2['Iy'] = (32*classlist2['F']**3)/(12*1000**4)
    classlist2['Mz'] = classlist2['H']*classlist2['I'] /1000
    classlist2['My'] = (classlist2['H']+40)*classlist2['J'] /1000
    #Calculating stresses for point a
    #Main equations that governs the bar
    # sigma = P/A - Mz*y/Iz - Myz/Iy
    # Tau = VzQ/ It
    classlist2['Sa'] = classlist2.apply(lambda x: 1000*x.K/x.Area + x.Mz*0.016/x.Iz, axis=1)
    classlist2['Tau_a'] = classlist2.apply(lambda x: (x.J*0.032*x.F/2*x.F/4)/(1000**2*x.Iz*0.032), axis=1)
    #For point b
    classlist2['Sb'] = classlist2.apply(lambda x: 1000*x.K/x.Area +x.Mz*0.016/x.Iz -x.My*-0.015/x.Iy, axis=1)
    classlist2['Tau_b'] = classlist2.apply(lambda x: (x.J*0.032*x.F/4*(x.F/4+x.F/8))/(1000**2*x.Iz*0.032), axis=1)
    #For point c
    classlist2['Sc'] = classlist2.apply(lambda x: 1000*x.K/x.Area +x.Mz*0.016/x.Iz -x.My*-0.03/x.Iy, axis=1)
    #As to thickness in the corner Qc is 0 therefore Tau_c is 0 for all
    #classlist2.iloc[student]
classlist2
```
