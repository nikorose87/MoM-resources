## Exercise 1

Two steel plates of uniform cross section 10 × 80 mm are welded together as shown. Knowing that centric 100-kN forces are applied to the welded plates and that β is your last code number times 5, determine (a) the in-plane shearing stress parallel to the weld, (b) the normal stress perpendicular to the weld. If your code is 0, please consider 25°, as an example if your code finishes in 8, your β will be 40°. 

![ima2](ex1.png)

```python
import numpy as np
import pandas as pd

classlist = pd.read_excel('classlist2021I.xlsx')
classlist = classlist.drop('Nivel', axis=1)

ex_numbers = classlist.copy()
ex_numbers['B'] = ex_numbers['ID'].apply(lambda x: x % 10 * 5)
ex_numbers['B'] = ex_numbers['B'].replace(0,25)
ex_numbers['TauMax'] = 62.5 * np.sin(np.deg2rad(2*ex_numbers['B']))
ex_numbers['SigmaRes'] = 62.5 + 62.5* np.cos(np.deg2rad(2*ex_numbers['B']))
ex_numbers
```

## Exercise 2

By means of the Mohr circle, determine two values of σy for which the maximum shearing stress is 90 MPa. Consider $\tau_{xy}= last \times 8$ and $\sigma_{x}= last^2 \times 5$  where last is your last code number, see the units in the figure. Consider 50 Mpa for both stresses if your code finishes in 0. 

![ima2](Ex2.png)

```python
ex2_numbers = classlist.copy()
ex2_numbers['tau'] = ex2_numbers['ID'].apply(lambda x: x % 10 * 8)
ex2_numbers['tau'] = ex2_numbers['tau'].replace(0,50)
ex2_numbers['sigma'] = ex2_numbers['ID'].apply(lambda x: ((x % 10) **2) *5)
ex2_numbers['sigma'] = ex2_numbers['sigma'].replace(0,50)
sigmay1 = lambda x,txy,tau: x-2*np.sqrt(tau**2-txy**2) 
sigmay2 = lambda x,txy,tau: x+2*np.sqrt(tau**2-txy**2) 
ex2_numbers['sigmay1'] = sigmay1(-ex2_numbers['sigma'], ex2_numbers['tau'], 90)
ex2_numbers['sigmay2'] = sigmay2(-ex2_numbers['sigma'], ex2_numbers['tau'], 90)
ex2_numbers
```
