## Exercise 1

Forces are applied at points A and B of the solid cast-iron bracket shown. Knowing that the bracket has a diameter of 0.8 in., determine the principal stresses and the maximum shearing stress at (a) point H, (b) point K.

![image1](8-41.png)

Please consider the following values for each:



```python
student = 19
import numpy as np
import pandas as pd
lock = False
if lock:
    classlist = pd.read_excel('classlist2021I.xlsx')
    classlist = classlist.drop('Nivel', axis=1)

    A = np.random.choice(np.arange(1500, 3500, 6),20) # lb
    B = np.random.choice(np.linspace(1.5, 3.5, 6),20) # in
    C = np.random.choice(np.arange(500, 1000, 6),20) # lb
    D = np.random.choice(np.linspace(2.0, 4.0, 6),20) # in
    E = np.random.choice(np.linspace(0.5, 1.5, 6),20) # in
    var_d = {'A':A,'B':B,'C':C,'D':D, 'E':E}

    for key, item in var_d.items():
        classlist[key] = np.random.choice(item,20)

    classlist = classlist.round(3)
    classlist.to_csv('problem1.csv')
else:
    classlist = pd.read_csv('problem1.csv')
```

```python
#Some common variables for all 
c = 0.4 #in
A = np.pi*c**2
I = (np.pi*c**4)/4
J = 2*I
Q = (c**3)*2/3
#Determining Mx
classlist['Mx'] = (classlist['D']-classlist['E'])*classlist['C']
#Determining Mz
classlist['Mz'] = -classlist['B']*classlist['C']


#At point H
classlist['SigmaH'] = -classlist['A']/A + classlist['Mx']*c/I
#Sigma average
classlist['SigmaAveH'] = classlist['SigmaH']/2
# Shear at Th
classlist['TauH'] = classlist['Mz']*c/J
# Radious
classlist['Rh'] = np.sqrt(classlist['SigmaAveH']**2 + classlist['TauH']**2)
# principal 1
classlist['sH1'] = classlist['SigmaAveH'] + classlist['Rh']
# principal 2
classlist['sH2'] = classlist['SigmaAveH'] - classlist['Rh']
# Tmax
classlist['TmaxH'] =  (classlist['sH1'] + classlist['sH2']) /2

# At point K
classlist['SigmaK'] = -classlist['A']/A
#Shear Tk
classlist['TauK'] = -classlist['Mz']*c/J + classlist['C']*Q/(I*2*c)
#Sigma average
classlist['SigmaAveK'] = classlist['SigmaK']/2
# Radious
classlist['Rk'] = np.sqrt(classlist['SigmaAveK']**2 + classlist['TauK']**2)
# principal 1
classlist['sK1'] = classlist['SigmaAveK'] + classlist['Rk']
# principal 2
classlist['sK2'] = classlist['SigmaAveK'] - classlist['Rk']
# Tmax
classlist['TmaxK'] =  (classlist['sK1'] - classlist['sK2']) /2

#classlist.iloc[student]
classlist
```
