### Carga axial por la influencia térmica

El rodillo AB contiene $n$ elementos, cada uno con una distribución homogénea en su sección transversal. El punto A se encuentra fijo, mientras que existe una luz $\delta_0$ cercana al punto B para quedar totalmente apoyada en sus dos secciones. Cada sección del rodillo contiene sus propiedades de longitud $L_i$, su diámetro $d$, y su módulo de elasticidad $E$ está dado por el material. Para el sistema general se brinda el $\delta_0$ [mm] y el $\Delta T$. [°C] Las unidades son metros, milímetros y GPa para $L_i$, $d_i$ y $E_i$, respectivamente.
